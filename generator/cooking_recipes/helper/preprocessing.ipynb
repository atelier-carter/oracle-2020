```python
with open('recipes.txt', 'r') as f:
    row = f.read()
```

```python
row = row.split('\n')
```

```python
with open('cooking_recipes.txt', 'a') as f: 
    for r in row:
        name = r.split('[')[0]
        f.write(name + '\n')
        f.write('\n')
        
        if len(r.split('[')) > 1:
            recipe = r.split('[')[1]
            ingredients = recipe.split(',')
            for ing in ingredients:
                f.write(ing + '\n')
            f.write('\n')
            f.write('\n')
            f.write('\n')
```
